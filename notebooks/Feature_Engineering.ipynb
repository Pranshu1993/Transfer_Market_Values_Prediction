```python
import pandas as pd
```

```python
df = pd.read_csv('/content/pre_processed_merged.csv')
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
df.head(10)
```

```python
df.columns
```

**Remove 'Kg', 'cm' from Weight & Height**

```python

df['Weight'] = df['Weight'].str.replace('kg',"")
df['Height'] = df['Height'].str.replace('cm',"")

df['Weight'] = df['Weight'].astype(float)
df['Height'] = df['Height'].astype(float)
```

```python
df.head(50)
```

**Applying Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
# Load the Excel file into a pandas dfFrame
# Create a LabelEncoder object
le = LabelEncoder()
for column in df.columns:
  if df[column].dtype == 'object' and column != 'Name':
    df[column] = le.fit_transform(df[column])
  #df.to_excel('file_encoded.xlsx', index=False) = X.head()
```

```python
df.head(50)
```

**Drop League**

```python
df = df.drop('league',axis=1)
```

```python
df.columns
```

**Split X Features & Y Features**

```python
X=df.drop(['Name','nationality','Market Value'],axis=1)
Y=df['Market Value']
```

**Feature Selection Techniques**

**1. Recursive Feature Selection**

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Perform recursive feature elimination with a linear regression model
estimator = LinearRegression()
selector = RFE(estimator, n_features_to_select=30, step=1)
selector.fit(X, Y)

# Print the selected features
print(X.columns[selector.support_])
```

**2. Lasso Regularisation L1 - Feature Selection** 

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.1)
lasso.fit(X,Y)

coef = pd.Series(lasso.coef_,index=X.columns)

# Print the top 30 features with the largest coefficient
top_features = coef.abs().sort_values(ascending=False)[:30]
print(top_features.index)
```

```python
a = ['minutes_played_overall', 'minutes_played_home', 'minutes_played_away',
       'appearances_overall', 'appearances_home', 'appearances_away',
       'goals_overall', 'goals_home', 'goals_away', 'assists_overall',
       'assists_home', 'assists_away', 'clean_sheets_overall',
       'clean_sheets_away', 'conceded_overall', 'conceded_home',
       'conceded_away', 'rank_in_league_top_midfielders', 'Skill Moves',
       'Crossing', 'Finishing', 'Short Passing', 'Volleys', 'Dribbling',
       'Sprint Speed', 'Stamina', 'Penalties', 'Sliding Tackle', 'GK Kicking',
       'GK Reflexes']

b = ['minutes_played_overall', 'appearances_overall', 'goals_overall',
       'conceded_overall', 'assists_overall', 'Sliding Tackle',
       'clean_sheets_overall', 'GK Reflexes', 'Crossing', 'Sprint Speed',
       'Short Passing', 'Finishing', 'min_per_match', 'Volleys',
       'rank_in_league_top_midfielders', 'GK Positioning',
       'rank_in_club_top_scorer', 'Stamina', 'Dribbling', 'penalty_goals',
       'Skill Moves', 'GK Kicking', 'clean_sheets_away', 'minutes_played_away',
       'Standing Tackle', 'GK Handling', 'Defense Awareness/Marking',
       'Penalties', 'Aggression', 'Year']
```

```python
in_first = set(a)
in_second = set(b)
in_second_but_not_in_first = in_second - in_first
Selected_features = a + list(in_second_but_not_in_first)

# len(Selected_features)
Selected_features.remove('Year')
len(Selected_features)
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
import joblib

# Scale the numerical columns using StandardScaler
scaler = StandardScaler()
X[Selected_features] = scaler.fit_transform(X[Selected_features])

# Save the scaler object as a joblib file
joblib.dump(scaler, 'scaler.joblib')

# Print the first 5 rows of the modified DataFrame
print(X.head())
```

```python
X['Market Value']= Y
```
