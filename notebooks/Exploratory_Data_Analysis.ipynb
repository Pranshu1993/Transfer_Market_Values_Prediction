```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm, skew

# Load the data into a Pandas DataFrame
df = pd.read_csv('/content/Merged_All_Data.csv')
df = df.drop(df.columns[0],axis=1)
df.rename(columns = {"Club_left":"Club","Name_left":"Name","Year_left":"Year"}, inplace = True)

df=df.drop(["best_match_score","birthday","word_overlap","overlap_count","birthday_GMT","Year_right","Name_right" ],axis=1)


for feature in df:
    # Check if the feature is numeric
    if df[feature].dtype != 'object':
        # Calculate the skewness of the feature
        skewness = skew(df[feature])
        
        # Plot the distribution of the feature
        sns.distplot(df[feature], fit=norm, kde=False, fit_kws={"color":"red", "label":"Normal Curve"}, hist_kws={"edgecolor":"k", "alpha":0.5})
        plt.axvline(df[feature].mean(), color='blue', linestyle='dashed', linewidth=2, label='Mean')
        plt.axvline(df[feature].median(), color='green', linestyle='dashed', linewidth=2, label='Median')
        plt.legend()
        plt.title(feature)
        plt.show()
        
        # Classify the skewness of the feature
df.to_csv("pre_processed_merged.csv",encoding="utf-8",index=False)
```
